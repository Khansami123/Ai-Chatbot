```python
import google.generativeai as genai
from IPython.display import Markdown, display

# Configure API key
GOOGLE_API_KEY = "AIzaSyDiTdhZYHWgcPV3xUEU85kAX7ZsOLYhCmI"
genai.configure(api_key=GOOGLE_API_KEY)
```

```python
system_instruction = """
You are a helpful and ethical AI assistant focused on defensive cybersecurity.
Always provide balanced and responsible advice. Do not assist with illegal activities.
"""

# Initialize the model with the system instruction
model = genai.GenerativeModel(
    model_name="gemini-2.5-flash-lite",
    system_instruction=system_instruction
)

# Start a chat session
chat_session = model.start_chat(history=[])

print("---CYBER SECURITY BOT ONLINE ---")
print("Type 'exit' to end the chat.")

while True:
    user_input = input("\n[You]: ")

    if user_input.lower() in ["exit", "quit", "stop"]:
        print("Bot: Goodbye!")
        break

    if "hack" in user_input.lower():
        print("I promote ethical cybersecurity only. I cannot assist with hacking.")
        continue


    try:
        response = chat_session.send_message(user_input)
        print("\n[Bot Response]:")
        display(Markdown(response.text))
        print("-" * 30)

    except Exception as e:
        print(f"An error occurred: {e}")
```
